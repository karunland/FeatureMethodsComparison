```python
import cv2 
import matplotlib.pyplot as plt
import os, sys
arg1 = sys.argv[1]

if (os.path.exists(f'./images/{arg1}')):
    os.remove(f'./images/{arg1}')
    img = cv2.imread(arg1)
else:
    print('File not found')
    exit()

print('Image shape:', img.shape)
print('Image data type:', img.dtype)

img = cv2.resize(img, (0,0), fx = 0.3, fy = 0.3)

plt.imshow(img)
plt.show()

# print image informations
print('Image shape:', img.shape)
print('Image data type:', img.dtype)


# save to file

cv2.imwrite(f'./images/{img}', img)
```
